# Definicja sterownika rozmytego

- Metoda wnioskowania: Mamdani
- T-norma: minimum (AND)
- S-norma: maximum (OR)
- Metoda defuzzyfikacji: Centroid
- Algorytm obliczania: KM

```python
controller = Mamdani(t_norm=min_t_norm, s_norm=max_s_norm, method="Centroid", algorithm="KM")
```

Wejścia i wyjścia:
- 5 wejść
- 1 wyjście

```python
# Wejścia
controller.add_input_variable('phq9_score')
controller.add_input_variable('bdi_score')
controller.add_input_variable('gad7_score')
controller.add_input_variable('sleep_quality')
controller.add_input_variable('social_interactions')

# Wyjście
controller.add_output_variable('depression_level')
```
